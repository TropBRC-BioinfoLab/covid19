```python
df= pd.read_csv('../latih_parted.csv', sep=",")
com_df=df['0']
prot_df=df['1282']

del(df['0'])
del(df['1282'])

bine=[]
for i in range(1,1283):
    bine.append(i)
df.columns=bine
das = df.astype(str)
nrow = das.shape[0]
ncol = das.shape[1]
with open('latih_libsvm', 'w') as f:
    for i in range(nrow) :
        dl = ''
        dl = dl + das.iloc[i][ncol]
        for j in range(1,ncol) :
            dl = dl + ' ' + str(j) + ':' + das.iloc[i][j]
        f.writelines(dl+'\n')
```
