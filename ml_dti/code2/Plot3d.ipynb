```python
# Author: Raghav RV <rvraghav93@gmail.com>
# License: BSD

import numpy as np
from matplotlib import pyplot as plt

from sklearn.datasets import make_hastie_10_2
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

print(__doc__)
```

```python
X, y = make_hastie_10_2(n_samples=8000, random_state=42)

# The scorers can be either be one of the predefined metric strings or a scorer
# callable, like the one returned by make_scorer
scoring = {'AUC': 'roc_auc', 'Accuracy': 'accuracy','F-measure':'f1','Precision':'precision', 'Recall':'recall'}

# Setting refit='AUC', refits an estimator on the whole dataset with the
# parameter setting that has the best cross-validated AUC score.
# That estimator is made available at ``gs.best_estimator_`` along with
# parameters like ``gs.best_score_``, ``gs.best_params_`` and
# ``gs.best_index_``
gs = GridSearchCV(DecisionTreeClassifier(random_state=42),
                  param_grid={'min_samples_split': range(2, 403, 10)},
                  scoring=scoring, refit='AUC', return_train_score=True)
gs.fit(X, y)
results = gs.cv_results_
```

```python
class uji(): 
    def __init__(self,data): 
        df0 = pd.read_csv('../data3/uji%snorm.csv'%data, delimiter= ",", header=None)
        df0=df0.fillna(0)
        self.com=df0[0]
        self.prot=df0[1282]
        self.x =df0.drop([0,1282], axis=1)
        df0[1300]=0
        
        self.y=df0[1300]
        self.dt=data
class latih(): 
    def __init__(self,data): 
        df = pd.read_csv("../data3/%s.csv"%data, delimiter=',')
        df.head()
        self.X = df.drop(['label'], axis=1)
        self.Y = df['label']
        self.dt=data
def dtuji(data): 
    return uji(data) 
def dtlatih(data): 
    return latih(data) 

df1= dtlatih('hasilsmote_1') #df1.X,df1.Y,df1.dt
df2= dtlatih('hasilsmote_2') #df2.X,df2.Y,df2.dt
df3= dtlatih('hasilsmote_3') #df3.X,df3.Y,df3.dt
df4= dtlatih('hasilsmote_4') #df4.X,df4.Y,df4.dt
df5= dtlatih('hasilsmote_5') #df5.X,df5.Y,df5.dt

dtf =dtuji('mbafika') #dtf.com, dtf.prot, dtf.x, dtf.y, dtf.dt
dtl =dtuji('mbalinda') #dtl.com, dtl.prot, dtl.x, dtl.y dtl.dt
```

```python
class lama(): 
    def __init__(self,data): 
        df0 = pd.read_csv('../data3/%s.csv'%data, delimiter= ",")
        df0=df0.fillna(0)
        self.com=df0['0']
        self.prot=df0['1282']
        self.X =df0.drop(['0','1282','label'], axis=1)
        self.Y = df0['label']
        self.dt=data
def dtlama(data): 
    return lama(data) 
do1 =dtlama('latih_parted_1') #dtf.com, dtf.prot, dtf.x, dtf.y, dtf.dt
do2 =dtlama('latih_parted_2')
do3 =dtlama('latih_parted_3')
do4 =dtlama('latih_parted_4')
do5 =dtlama('latih_parted_5')
```

```python
def plot2d(d):
    dt= dtlatih('hasilsmote_{0}'.format(d))
    dl =dtlama('latih_parted_{0}'.format(d))
    
    pipeline = Pipeline([('scaling', Normalizer()), ('pca', PCA(n_components=3))])
    
    pca_result2 = pipeline.fit_transform(dl.X.values)
    pca_result2=pd.DataFrame(pca_result2)
    pca_result2=pd.concat([pca_result2,dl.Y],axis=1)
    X_new = pca.inverse_transform(pca_result2)
    fig = plt.subplots(figsize=(8,4))
    plt.scatter(X_new[:, 0], X_new[:, 1],X_new[:, 2], alpha=0.8,label=X_new[:, 3])
    plt.axis('equal');
    
    fig = plt.subplots(figsize=(8,4))
    pca_result = pipeline.fit_transform(dt.X.values)
    X_new = pca.inverse_transform(pca_result)
    plt.scatter(X_new[:, 0], X_new[:, 1],X_new[:, 2], alpha=0.8)
    plt.axis('equal');
    
```

```python
def plotsample3d(df1):
    sns.set_style("white")
    my_dpi=360
    plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)
    df1p=df1.X[df1.Y==1]
    df1n=df1.X[df1.Y==0]
    #df1.Y=pd.DataFrame(df1.Y)
    df1.Y=pd.Categorical(df1.Y)
    my_color=df1.Y.codes
    df1p.Y=pd.Categorical(df1.Y[df1.Y==1])
    my_colorp=df1p.Y.codes
    df1n.Y=pd.Categorical(df1.Y[df1.Y==0])
    my_colorn=df1n.Y.codes
    pca = PCA(n_components=3)
    pca.fit(df1.X)
    pcap = PCA(n_components=3)
    pcap.fit(df1p)
    pcan = PCA(n_components=3)
    pcan.fit(df1n)
    result=pd.DataFrame(pca.transform(df1.X), columns=['PCA%i' % i for i in range(3)], index=df1.X.index)
    resultp=pd.DataFrame(pcap.transform(df1p), columns=['PCA%i' % i for i in range(3)], index=df1p.index)
    resultn=pd.DataFrame(pcan.transform(df1n), columns=['PCA%i' % i for i in range(3)], index=df1n.index)
    # Plot initialisation
    fig = plt.figure(figsize=(16,10))
    ax = fig.add_subplot(111, projection='3d')

    #sc=ax.scatter(result['PCA0'], result['PCA1'], result['PCA2'], c=my_color, cmap="Blues", s=10)
    scp=ax.scatter(resultp['PCA0'], resultp['PCA1'], resultp['PCA2'], c=my_colorp, label='positif',cmap="RdYlBu_r", s=30)
    scn=ax.scatter(resultn['PCA0'], resultn['PCA1'], resultn['PCA2'], c=my_colorn,label='negatif', cmap="Reds_r", s=30)

    # make simple, bare axis lines through space:
    xAxisLine = ((min(result['PCA0']), max(result['PCA0'])), (0, 0), (0,0))
    ax.plot(xAxisLine[0], xAxisLine[1], xAxisLine[2], 'r')
    yAxisLine = ((0, 0), (min(result['PCA1']), max(result['PCA1'])), (0,0))
    ax.plot(yAxisLine[0], yAxisLine[1], yAxisLine[2], 'r')
    zAxisLine = ((0, 0), (0,0), (min(result['PCA2']), max(result['PCA2'])))
    ax.plot(zAxisLine[0], zAxisLine[1], zAxisLine[2], 'r')

    # label the axes
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.set_zlabel("PC3")
    #ax.set_title("PCA on the smote dataset")
    colors = [sc.cmap(sc.norm(i)) for i in [1, 2, 0]]
    custom_lines = [plt.Line2D([],[], ls="", marker='.', 
                    mec='k', mfc=c, mew=.1, ms=20) for c in colors]
    
    ax.legend()
    plt.show()
    fig.savefig('../data3/{0}.png'.format(df1.dt), bbox_inches='tight')
```
