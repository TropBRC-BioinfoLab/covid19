```python
import pubchempy as pcp
import pandas as pd
import numpy as np
import re, os
```

```python
dti=pd.read_excel('drug_tambahan_2.xlsx',header=None)  
dti.head()
```

```python
listIdCom = dti[0]
listIdCom = [str(x) for x in listIdCom]
```

```python
def split(word): 
    return [char for char in word]  
```

```python
len(dti)
```

```python
df_= np.zeros((881,len(dti)))
df_=pd.DataFrame(df_)
```

```python
for i in range(0,len(dti)-1):
    df_[i][0]=listIdCom[i]
    com = pcp.Compound.from_cid(listIdCom[i]).cactvs_fingerprint
    split_strings = split(com)
    df_[i] = pd.DataFrame(split_strings)
df_.head()
```

```python
df_=df_.T
df_.tail()
```

```python
df_=pd.concat([dti[1],df_], axis=1)
df_.tail()
```

```python
df_.to_csv(r'fp_tambahan_2.csv', index=None, header=None)
```
